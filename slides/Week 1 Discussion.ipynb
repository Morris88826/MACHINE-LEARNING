### Introduction to Numerical Python

When dealing with numerical computations in `Python`, the `numpy` package is widely used. We will get familiar with some of its fundamental functionalities here.

To utilize `numpy`, the initial step is to import it:

```python
import numpy as np
```

Note that we've assigned the `numpy` module the alias `np` for the sake of convenient reference.

In numpy, the term *array* is a generic reference to a multidimensional collection of numbers. The `np.array()` function is employed to define arrays. Here we use it to create one-dimensional arrays referred to as *vectors*, such as `x` and `y`. The syntax `np.array()` signifies that the invoked function is from the `numpy` package.

```python
x = np.array([3, 4, 5])
y = np.array([4, 9, 7])
```

As `x` and `y` have been defined using `np.array()`, the addition of these arrays yields a meaningful result. This is in contrast to our previous results in the earlier section when attempting to add two lists without utilizing `numpy`.

```python
x + y
```

In `numpy`, matrices are commonly represented as two-dimensional arrays, while vectors are represented as one-dimensional arrays. To create a two-dimensional array, we can use the following approach:

```python
x = np.array([[1, 2], [3, 4]])
x
```

The object `x` possesses various attributes. To access an attribute of `x`, the syntax involves typing `x.attribute`, replacing "attribute" with the actual attribute name. For example, to access the `ndim` attribute of `x`, the notation is as follows:

```python
x.ndim
```

The output reveals that `x` is a two-dimensional array. Similarly, `x.dtype` represents the data type attribute of the object `x`, indicating that `x` is composed of 64-bit integers:

```python
x.dtype
```

The reason `x` consists of integers is that we created it by exclusively providing integers to the `np.array()` function. If we had included decimals in the input, the result would have been an array of floating-point numbers:

```python
np.array([[1, 2], [3.0, 4]]).dtype
```

As seen above, we can type `fun?` to see the documentation of the function  

As demonstrated earlier, typing `fun?` allows us to access the documentation for the function `fun`. Similarly, applying this approach to the function `np.array()` reveals that we can create a floating-point array by including a `dtype` argument in the `np.array()` function call:

```python
np.array?
```

```python
np.array([[1, 2], [3, 4]], float).dtype
```

The array `x` is two-dimensional. To determine the number of rows and columns, you can inspect its `shape` attribute.

```python
x.shape
```

A method is a function linked to an object. For example, with an array object `x`, the expression `x.sum()` computes the sum of all its elements using the `sum()` method designed for arrays. The invocation `x.sum()` automatically supplies `x` as the first argument to its `sum()` method.

```python
x = np.array([1, 2, 3, 4])
x.sum()
```

Alternatively, we could sum the elements of `x` by passing `x` as an argument to the `np.sum()` function.

```python
x = np.array([1, 2, 3, 4])
np.sum(x)
```

As another illustration, the `reshape()` method generates a new array with identical elements to `x` but with a different shape. This is accomplished by providing a tuple, in this case, `(2, 3)`, as an argument to `reshape()`. The tuple indicates the creation of a two-dimensional array with `2` rows and `3` columns.

```python
x = np.array([1, 2, 3, 4, 5, 6])
print('beginning x:')
print(x)
print('')
x_reshape = x.reshape((2, 3))
print('reshaped x:')
print(x_reshape)
```

The preceding output highlights that numpy arrays are specified as a sequence of *rows*. This ordering is referred to as *row-major ordering*, distinguishing it from *column-major ordering*.

`Python` (and consequently `numpy`) employs 0-based indexing. Therefore, to retrieve the top-left element of `x_reshape`, the notation used is `x_reshape[0, 0]`.

```python
x_reshape[0, 0]
```

Likewise, `x_reshape[1, 2]` provides the element in the second row and the third column of `x_reshape`.

```python
x_reshape[1, 2]
```

In a one-dimensional scenario, a single index suffices for accessing elements. For instance, `x[2]` retrieves the third entry of `x`.

Now, let's modify the top-left element of `x_reshape`. Interestingly, we find that the first element of `x` has also been altered:

```python
print('x before we modify x_reshape:')
print (x)
print ('')

print('x_reshape before we modify x_reshape:')
print(x_reshape)
print ('')

x_reshape[0, 0] = 5

print('x_reshape after we modify its top left element:')
print(x_reshape)
print ('')

print('x after we modify top left element of x_reshape:')
print(x)
```

The modification of `x_reshape` also impacted `x` because the two objects share the same memory space.

Now, let's briefly touch upon some attributes of arrays that will be useful. An array's shape attribute holds its dimensions, represented as a tuple. The `ndim` attribute provides the number of dimensions, while `T` yields its transpose.

```python
x_reshape.shape , x_reshape.ndim , x_reshape.T
```

```python
np.sqrt(x)
```

Moreover, squaring the elements can be achieved with the following expression:

There will be instances where generating random data is required. The `np.random.normal()` function serves to generate a vector of random normal variables. To acquire more information about this function, you can refer to the help page by executing `np.random.normal?`. The initial line of the help page provides the signature: `normal (loc=0.0, scale=1.0, size=None)`. This indicates that the function takes arguments such as `loc`, `scale`, and `size`, which are keyword arguments. This implies that when these arguments are supplied to the function, they can be referenced by name, and the order of input does not matter. By default, the function generates random normal variables with a mean (`loc`) of 0 and a standard deviation (`scale`) of 1. Additionally, a single random variable is generated unless the `size` argument is modified.

Now, let's generate 50 independent random variables from a normal distribution with mean 0 and standard deviation 1 ($N(0, 1)$).

```python
x = np.random.normal(size=50)
x
```

We can create an array, `y`, by adding an independent $N(50, 1)$ random variable to each element of `x`.

```python
y = x + np.random.normal(loc=50, scale=1, size=50)
y
```

The `np.corrcoef()` function calculates the correlation matrix between `x` and `y`. The off-diagonal elements of the matrix provide the correlation between `x` and `y`.

```python
np.corrcoef(x, y)
```

If you run the above commands again, you'll likely observe a different set of results each time. Specifically, each invocation of `np.random.normal()` produces a different outcome, as demonstrated in the following example.

```python
print(np.random.normal(scale=5, size=2))
print(np.random.normal(scale=5, size=2))
```

To guarantee consistent results with each run of our code, we can establish a random seed using the `np.random.default_rng()` function. This function requires an arbitrary integer argument specified by the user. By setting a random seed before generating random data, subsequent runs of our code will produce identical results. The object rng encompasses essentially all the random number generation methods found in `np.random`. Consequently, to generate normal data, we use `rng.normal()`.

```python
random_seed = 1303
rng = np.random.default_rng (random_seed)
print(rng.normal(scale=5, size=2))
rng2 = np.random.default_rng (random_seed)
print(rng2.normal(scale=5, size=2))
```

```python
print(rng.normal(scale=1, size=x.shape))
```

For reproducibility, we use `np.random.default_rng()` whenever calculations involving random quantities within numpy are performed. However, it's worth noting that with the release of new `numpy` versions, there is a possibility of minor discrepancies between the output from `numpy`.

To create a data frame consisting of the `'weight'` and `'origin'` of the subset of cars with a year greater than 80 (those built after 1980), we start by creating a Boolean array that indexes the rows. The `loc[]` method supports both Boolean entries and strings for this purpose:

```python
idx_80 = Auto_re['year'] > 80
Auto_re.loc[idx_80 , ['weight', 'origin']]
```

or more concicely:

```python
Auto_re.loc[lambda df: df['year'] > 80, ['weight', 'origin']]
```

The lambda call generates a function that takes a single argument, in this case, df, and returns `df['year']>80`. Since it is created within the `loc[]` method for the dataframe `Auto_re`, that dataframe will be the supplied argument.

As another example of using a `lambda`, let's consider wanting all cars built after 1980 that achieve greater than 30 miles per gallon:

```python
Auto_re.loc[lambda df: (df['year'] > 80) & (df['mpg'] > 30),
['weight', 'origin']
]
```

The symbols `&` and `|` perform element-wise AND and element-wise OR operations, respectively. For instance, if we want to retrieve all Ford and Datsun cars with a displacement less than 300, we check whether each `'name'` entry contains either the string 'ford' or 'datsun' using the `str.contains()` method of the index attribute of the dataframe:

```python
Auto_re.loc[lambda df: (df['displacement'] < 300)
& (df.index.str.contains('ford')
| df.index.str.contains('datsun')),
['weight', 'origin']
]
```

### Graphics

In `Python`, it's common practice to utilize the `matplotlib` library for graphics. However, since `Python` was not originally designed with data analysis in mind, the concept of plotting is not inherent to the language. To create a figure and the axes for plotting our data, we will use the `subplots()` function from `matplotlib.pyplot`.

In `matplotlib`, a plot comprises a `figure` and one or more `axes`. The `figure` can be envisioned as a blank canvas upon which one or more plots will be displayed—it represents the entire plotting window. The `axes`, on the other hand, encompass crucial information about each plot, including its x- and y-axis labels, title, and more. It's important to note that in `matplotlib`, the term *axes* does not represent the plural of "axis." The plot's axes contain more information than just the x-axis and y-axis.

To start, we import the `subplots()` function from `matplotlib`, which we use consistently when creating figures. This function returns a tuple of length two, consisting of a `figure` object and the corresponding `axes` object. Usually, we pass the `figsize` parameter as a keyword argument. After creating our `axes`, we initiate our initial plot using its `plot()` method.

```python
from matplotlib.pyplot import subplots
fig , ax = subplots(figsize=(8, 8))
x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x, y);
```

Take note that we have unpacked the tuple of length two, which is returned by the `subplots()` function, into the two distinct variables `fig` and `ax`. Unpacking is generally favored over the following equivalent but slightly more verbose code:

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html
output = subplots(figsize=(8, 8))
fig = output[0]
ax = output[1]
```

Alternatively we can create a scatterplot with custom markers:

```python
fig , ax = subplots(figsize=(8, 8))
ax.scatter(x, y, marker='o');
```

For labeling our plot, we utilize the `set_xlabel()`, `set_ylabel()`, and `set_title()` methods of the `ax` variable.

```python
fig, ax = subplots(figsize=(8, 8))
ax.scatter(x, y, marker='o')
ax.set_xlabel("this is the x-axis")
ax.set_ylabel("this is the y-axis")
ax.set_title("Plot of X vs Y");
```

At times, we may need to generate multiple plots within a single figure. This can be accomplished by providing additional arguments to the `subplots()` function. In the example below, we create a $2 × 3$ grid of plots in a figure whose size is determined by the `figsize` argument. In such scenarios, there is often a connection between the axes in the plots. For instance, all plots may share a common x-axis. The `subplots()` function can automatically manage this situation by using the keyword argument `sharex=True`. The `axes` object below is an array that points to different plots within the figure.

```python
fig , axes = subplots(
    nrows=2,
    ncols=3,
    figsize=(15, 5))
```

Now, we create a scatter plot with *'o'* in the second column of the first row and another scatter plot with *'+'* in the third column of the second row.

```python
axes[0,1].plot(x, y, 'o')
axes[1,2].scatter(x, y, marker='+')
fig
```

To save the output of the `fig`, we invoke its `savefig()` method. The dpi argument in `.savefig()` represents dots per inch, determining the resolution and size of the figure in pixels.

```python
fig.savefig("Figure.png", dpi=400)
fig.savefig("Figure.pdf", dpi=200);
```

We have the flexibility to make step-by-step updates to `fig`. For instance, we can adjust the range of the x-axis, re-save the figure, and even re-display it.

```python
axes[0,1].set_xlim([-1,1])
fig.savefig("Figure_updated.jpg")
fig
```

Now, let's generate more sophisticated plots. The `ax.contour()` method is used to create a contour plot, which represents three-dimensional data akin to a topographical map. It requires three arguments:

A vector of `x` values (representing the first dimension),
A vector of `y` values (representing the second dimension), and
A matrix whose elements correspond to the `z` value (representing the third dimension) for each pair of `(x, y)` coordinates.
To create `x` and `y`, we'll utilize the command `np.linspace(a, b, n)`, which generates a vector of `n` numbers starting at `a` and ending at `b`.

```python
# We plot what does linspace produce
fig , ax = subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.plot(f);
print(f"{(1 / (1 + x**2)).shape}, {np.cos(y).shape}")
```

```python
fig , ax = subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f);
```

```python
fig , ax = subplots(figsize=(8, 8))
ax.contour(x, y, f, levels=45);
```

The `ax.imshow()` method is akin to `ax.contour()`, with the difference that it generates a color-coded plot where colors are determined by the `z` value. This type of plot is commonly referred to as a heatmap and is occasionally employed to visualize temperature in weather forecasts.

```python
fig , ax = subplots(figsize=(8, 8))
ax.imshow(f);
```
